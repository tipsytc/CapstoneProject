```python
import pandas as pd
import numpy as np
import random
```

```python
df_base = pd.read_csv('BASE202010.csv', low_memory = False, index_col=[0])
```

```python
df_base[df_base.MRR_REV == 0]
```

```python
#Here we add a random fluctuation on 4 columns,
#note that the fluctuation of REV columns do not
#corelate with the DATA_MB

#Just for demonstration purpose

def new_fake(df, columns):
    rand = np.array([random.uniform(0.8,1.2) for i in range(df_base.shape[0])])
    for column in columns:
        df[column] = df[column] * rand
    new_df = df.iloc[np.random.randint(625776, size=625000)]
    return new_df
```

```python
a = new_fake(df_base, ['MRR_REV', 'TOTAL_REV', 'MSF_REV', 'DATA_MB'])
```

```python
for i in range(2,10):
    s = 'BASE_20200' + str(i) + '.csv'
    fake_df = new_fake(df_base, ['MRR_REV', 'TOTAL_REV', 'MSF_REV', 'DATA_MB'])
    fake_df.to_csv(s, index=False)  
```
